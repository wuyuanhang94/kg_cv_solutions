```python
import matplotlib.pyplot as plt
import pandas as pd
metrics_df = pd.read_json("/home/yiw/kg/sart/output/metrics.json", orient="records", lines=True)
mdf = metrics_df.sort_values("iteration")
# metrics_df

fig, ax = plt.subplots()

mdf1 = mdf[~mdf["MaP IoU"].isna()]
# mdf1

ax.plot(mdf1["iteration"], mdf1["MaP IoU"], c="C0", label="train")
ax.grid("on")
ax.legend()
ax.set_title("MaP IoU")
plt.show()
```
